```python
    

#Q3. What is App routing in Flask? Why do we use app routes?

"""
app routing is basiclly according their name route or path which means ;define a path for client to 
reaching at a server.It make the keypoint for url, by which a user can find that specific url.Example 
are given below and here can acsses any function wrapped in @app.route method. we can esily access the
welcome function by given /rohan in the end of
the url.
"""
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

@app.route("/rohan")
def welcome():
    return "Rohan ! most welcome in this world"

if __name__=="__main__":
    app.run(host="0.0.0.0")

    
    
```
